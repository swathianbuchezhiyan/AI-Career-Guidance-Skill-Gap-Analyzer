```python
import pandas as pd

df = pd.read_csv("career_data.csv")

df.head()
```

```python
df.info()
df.isnull().sum()
df.duplicated().sum()
```

```python
df.describe(include='all')
```

```python
# User input
user_skills = input("Enter your skills (comma separated): ").lower().split(",")
user_skills = [skill.strip() for skill in user_skills]

results = []

# Process each role
for index, row in df.iterrows():

    role = row["Role"]

    # Convert required skills to list
    required_skills = [skill.strip().lower() for skill in row["Required_Skills"].split(",")]

    # Matching skills
    matched = [skill for skill in user_skills if skill in required_skills]
    score = len(matched)

    # Missing skills
    missing = [skill for skill in required_skills if skill not in user_skills]

    # Match percentage
    if len(required_skills) > 0:
        match_percent = (score / len(required_skills)) * 100
    else:
        match_percent = 0

    # Store results
    results.append([role, score, round(match_percent, 2), missing])

# Sort by best match
results.sort(key=lambda x: x[1], reverse=True)

# Output Top 3
print("\n🔥 Top 3 Career Recommendations:\n")

for i in results[:3]:
    print("Role:", i[0])
    print("Match Score:", i[1])
    print("Match %:", i[2], "%")
    print("Missing Skills:", ", ".join(i[3]))
    print("-" * 40)

# Best recommendation
best = results[0]

print("\n🏆 Best Career Suggestion:", best[0])
print("Match %:", best[2], "%")
```
